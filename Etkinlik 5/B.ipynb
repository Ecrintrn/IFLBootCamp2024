```python
import math

def cift_sayi_mi(liste, orta, kucuk=True):
    for i in liste:
        if i % 2 != 0:
            return False
    cift = cift_sirali_mi(liste, orta, kucuk)
    return cift

def tek_sayi_mi(liste, orta, kucuk=True):
    for i in liste:
        if i % 2 == 0:
            return False
    tek = cift_sirali_mi(liste, orta, kucuk)
    return tek
    
def cift_sirali_mi(liste, orta, kucuk=True):
    num = liste[0]
    if orta == "orta1":
        if num + 4 != liste[1]:
            return False
        num += 4
        if num + 2 != liste[2]:
            return False
        return True
    elif orta == "orta2":
        if num + 2 != liste[1]:

            return False
        num += 2
        if num + 4 != liste[2]:
            return False
        return True
    else:
        num = liste[0]
        for i in liste[1:]:
            if num+2 != i:
                return False
            num = i
        return True
        
def ciftTek_hesaplama(liste, orta, kucuk=True):
    if kucuk == True and orta == "orta1":
        return (liste[0] + 2)
    elif kucuk == True and orta == "orta2":
        return (liste[1] + 2)
    elif kucuk == False:
        return (liste[0] - 2)
    return liste[2] + 2
    
def asal_sayi_hesaplama(liste, orta, kucuk=True):
    ters = False
    if kucuk == True and orta == "orta1":
        return sonraki_asal(liste[0], ters)
    elif kucuk == True and orta == "orta2":
        return sonraki_asal(liste[1], ters)
    elif kucuk == False:
        return sonraki_asal(liste[0], ters=True)
    return sonraki_asal(liste[2], ters)

def sonraki_asal(sayi, ters):
    if ters ==False:
        sayi += 1
        while True:
            asal = True
            for i in range(2, int(math.sqrt(sayi)) + 1):
                if sayi % i == 0:
                    asal = False
                    break
            if asal:
                return sayi
            sayi += 1
    else:
        sayi -= 1
        while True:
            asal = True
            for i in range(int(math.sqrt(sayi)) + 1, 2):
                if sayi % i == 0:
                    asal = False
                    break
            if asal:
                return sayi
            sayi += 1

def asal_sayi_mi(liste, orta, kucuk=True):
    for i in liste:
        if i < 2:
            return False
        for k in range(2, int(math.sqrt(i)) + 1):
            if i % k == 0:
                return False
    return asal_sirali_mi(liste, orta, kucuk)

def asal_sirali_mi(liste, orta, kucuk=True):
    asal = asal_liste(liste, orta, kucuk)
    if orta == "orta1":
        if liste[0] == asal[0] and liste[1] == asal[2] and liste[2] == asal[3]:
            return True
    elif orta == "orta2":
        if liste[0] == asal[0] and liste[1] == asal[1] and liste[2] == asal[3]:
            return True
    elif orta == "kenar":
        for i in range(len(liste)):
            if liste[i] != asal[i]:
                return False
        return True
    return False

def asal_liste(liste, orta, kucuk=True):
    alt = liste[0]
    ust = liste[2]
    asal = []
    for sayi in range(alt,ust + 1):  
       if sayi > 1:  
           for i in range(2,sayi):  
               if (sayi % i) == 0:  
                   break  
           else:  
               asal.append(sayi)
    return asal

def fibonacci_mi(liste, orta, kucuk=True):
    for i in liste:
        if not fibonacci_kontrol(i):
            return False, 0
    liste, sayi = fibonacci_hesaplama(liste, orta, kucuk)
    fibonacci = fibonacci_sirali_mi(liste, orta, kucuk)
    if fibonacci:
        return True, sayi
    return False, sayi

def fibonacci_kontrol(n):
    return karekök_hesaplama(5 * (n * n) + 4) or karekök_hesaplama(5 * (n * n) - 4)

def fibonacci_sirali_mi(liste, orta, kucuk=True):
    if liste[0] + liste[1] == liste[2]:
        return True
    return False

def fibonacci_hesaplama(liste, orta, kucuk=True):
    if kucuk == True and orta == "orta1":
        sayi = liste[2] - liste[1]
        liste.insert(1, sayi)
    elif kucuk == True and orta == "orta2":
        sayi = liste[0] + liste[1]
        liste.insert(2, sayi)
    elif kucuk == False:
        sayi = liste[1] - liste[0]
        liste.insert(0, sayi)
    else:
        sayi = liste[1] + liste[2]
        liste.append(sayi)
    return liste, sayi

def karesel_sayi_dizisi(liste, orta, kucuk=True):
    for i in liste:
        if not karekök_hesaplama(i):
            return False, 0
    liste, sayi = karesel_sayi_hesaplama(liste, orta, kucuk)
    karesel = karesel_sirali_mi(liste, kucuk)
    if karesel:
        return True, sayi
    return False, sayi

def karekök_hesaplama(num):
    return int(math.sqrt(num)) ** 2 == num

def karesel_sayi_hesaplama(liste, orta, kucuk=True):
    if kucuk == True and orta == "orta1":
        sayi = int(math.sqrt(liste[1])) - 1
        liste.insert(1, sayi**2)
    elif kucuk == True and orta == "orta2":
        sayi = int(math.sqrt(liste[1])) + 1
        liste.insert(2, sayi**2)
    elif kucuk == False:
        sayi = int(math.sqrt(liste[0])) - 1
        liste.insert(0, sayi**2)
    else:
        sayi = int(math.sqrt(liste[2])) + 1
        liste.append(sayi**2)
    return liste, sayi

def karesel_sirali_mi(liste, kucuk=True):
    if int(math.sqrt(liste[0])) + 2 ==  int(math.sqrt(liste[1])) + 1  == int(math.sqrt(liste[2])):
        return True
    return False

def palindrom_sayi_mi(sayi):
    sayi_str = str(sayi)
    return sayi_str == sayi_str[::-1]

def palindrom_sayi_hesaplama(liste, kucuk=True):
    if kucuk == False:
        for i in range(liste[0]-1, liste[0] - 100, -1):
            if palindrom_sayi_mi(i):
                return i
    for i in range(liste[2] + 1, liste[2] + 100):
        if palindrom_sayi_mi(i):
            return i  

def palindrom_sirali_mi(liste):
    for sayi in liste:
        if not palindrom_sayi_mi(sayi):
            return False
    return True

def ucgensel_dizi_mi(liste, orta, kucuk=True):
    if kucuk == True and orta == "orta2":
        sayi = ((liste[1] - liste[0]) +1) + liste[1]
        liste.insert(2, sayi)
    elif kucuk == True and orta == "orta1":
        sayi = abs(liste[1] - ((liste[2] - liste[1]) -1))
        liste.insert(1, sayi)
    elif kucuk == False:
        sayi = abs(liste[0] - ((liste[1] - liste[0]) -1))
        liste.insert(0, sayi)
    else:
        sayi = ((liste[2] - liste[1]) +1) + liste[2]
        liste.append(sayi)
    ucgensel = ucgensel_sirali_mi(liste, kucuk)
    if ucgensel:
        return True, sayi
    return False, sayi

def ucgensel_sirali_mi(liste, kucuk=True):
    num = liste[0]
    for i in liste[1:]:
        if i != ucgensel_mi(num, i, kucuk):
            return False
        num = i
    return True

def ucgensel_mi(num, i, kucuk=True):
    ucgen = 0
    for x in range(0, i + 1):
            if x**2 + x == 2 * i:
                ucgen = x
    num += ucgen
    return num

def aritmetrik_dizi(liste):
    if liste[2] - liste[1] == liste[1] - liste[0]:
        return ( liste[1] - liste[0]) + liste[2]

def geometrik_dizi(liste):
    if liste[2] / liste[1] == liste[1] / liste[0]:
        return int((liste[1] / liste[0]) * liste[2])


def bulunmayan_sayi(liste):
    i = 0
    aranan_sayi = None
    
    while i < len(liste):
        if liste[i] == '?':
            aranan_sayi = i
            break
        i += 1

    orta1 = False
    orta2 = False
    if aranan_sayi == 0:
        s1, s2, s3 = int(liste[1]), int(liste[2]), int(liste[3])
        liste = [s1, s2, s3]
    elif aranan_sayi == 1:
        orta1 = True
        s1, s2, s3 = int(liste[0]), int(liste[2]), int(liste[3])
        liste = [s1, s2, s3]
    elif aranan_sayi == 2:
        orta2 = True
        s1, s2, s3 = int(liste[0]), int(liste[1]), int(liste[3])
        liste = [s1, s2, s3]
    elif aranan_sayi == 3:
        s1, s2, s3 = int(liste[0]), int(liste[1]), int(liste[2])
        liste = [s1, s2, s3]

    if orta1:
        return "orta1", aranan_sayi, liste
    elif orta2:
        return "orta2", aranan_sayi, liste
    else:
        return "kenar", aranan_sayi, liste
def hepsi():
    while True:
        liste = input("Örüntünün 3 tane sayısını sırayla giriniz: ").split(", ")
        if len(liste) == 4:
            orta, aranan_sayi, liste = bulunmayan_sayi(liste)
        
        kucuk = True
        if orta == "kenar":
            if liste[0]>liste[2] :
                liste = liste[::-1]
                kucuk = False
            else:
                liste
            if aranan_sayi == 0:
                kucuk = False
        
        cift_mi = cift_sayi_mi(liste, orta, kucuk)
        tek_mi = tek_sayi_mi(liste, orta, kucuk)
        asal_mi = asal_sayi_mi(liste, orta, kucuk)
        fibonacci, sayi = fibonacci_mi(liste, orta, kucuk)
        karesel_mi, karesel = karesel_sayi_dizisi(liste, orta, kucuk)
        ucgensel, ucgen = ucgensel_dizi_mi(liste, orta, kucuk)
        
        #palindrom = palindrom_sayi_mi(liste, kucuk, orta)
        #aritmetrik = aritmetrik_dizi(liste, kucuk, orta)
        #geometrik = geometrik_dizi(liste, kucuk, orta)
        
        print(f"Çift Sayı Dizisi: {ciftTek_hesaplama(liste, orta, kucuk) if cift_mi else '-'}")
        print(f"Tek Sayı Dizisi: {ciftTek_hesaplama(liste, orta, kucuk) if tek_mi else '-'}")
        print(f"Asal Sayı Dizisi: {asal_sayi_hesaplama(liste, orta, kucuk) if asal_mi else '-'}")
        print(f"Fibonacci Sayı Dizisi: {sayi if fibonacci else '-'}")
        print(f"Karesel Sayı Dizisi: {karesel**2 if karesel_mi else '-'}")
        print(f"Uçgensel Sayı Dizisi: {ucgen if ucgensel else '-'}")
        #print(f"Aritmetrik Sayı Dizisi: {aritmetrik if aritmetrik else '-'}")
        #print(f"Geometrik Sayı Dizisi: {geometrik if geometrik else '-'}")
        baska = input("Başka örüntü inceleyecek misiniz? [e/h] : ")
        if baska.lower() != "e":
            print("Matematik örüntüleri kapsamında çalışmamız sona ermiştir. Güle güle… :)")
            break

hepsi()
```
