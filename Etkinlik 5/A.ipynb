```python
import math

def cift_sayi_mi(liste):
    for i in liste:
        if i % 2 != 0:
            return False
    cift = cift_sirali_mi(liste)
    return cift

def cift_sirali_mi(liste):
    num = liste[0]
    for i in liste[1:]:
        if num+2 != i:
            return False
        num = i
    return True

def tek_sayi_mi(liste):
    for i in liste:
        if i % 2 == 0:
            return False
    tek = cift_sirali_mi(liste)
    return tek

def asal_sayi_hesaplama(liste):
    num = liste[2] + 1
    while True:
        if asal_sayi_mi([num]):
            return num
        num += 1

def asal_sayi_mi(liste):
    for i in liste:
        if i < 2:
            return False
        for k in range(2, int(math.sqrt(i)) + 1):
            if i % k == 0:
                return False
    return asal_sirali_mi(liste)

def asal_sirali_mi(liste):
    num = liste[0]
    for i in liste[1:]:
        if i != asal_mi(num, liste, liste.index(i)):
            return False
        num = i 
    return True

def asal_mi(num, liste, index):
    n = num + 1
    while True:
        is_prime = True
        for k in range(2, int(math.sqrt(liste[index])) + 1):
            if n % k == 0:
                is_prime = False
                break
        if is_prime:
            return n
        else:
            n+=1

def fibonacci_mi(liste):
    for i in liste:
        if not fibonacci_kontrol(i):
            return False
    fibonacci = fibonacci_sirali_mi(liste)
    if fibonacci:
        return True
    return False

def fibonacci_kontrol(n):
    return karekök_hesaplama(5 * (n * n) + 4) or karekök_hesaplama(5 * (n * n) - 4)

def fibonacci_sirali_mi(liste):
    if liste[0] + liste[1] == liste[2]:
        return True
    return False

def fibonacci_hesaplama(liste):
    return liste[1] + liste[2]

def karesel_sayi_dizisi(liste):
    for i in liste:
        if not karekök_hesaplama(i):
            return False
    karesel = karesel_sirali_mi(liste)
    if karesel:
        return True
    return False

def karekök_hesaplama(num):
    return int(math.sqrt(num)) ** 2 == num

def karesel_sayi_hesaplama(liste):
    ücüncüSayi = int(math.sqrt(liste[2]))
    return (ücüncüSayi+1) ** 2

def karesel_sirali_mi(liste):
    if int(math.sqrt(liste[0])) + 2 ==  int(math.sqrt(liste[1])) + 1  == int(math.sqrt(liste[2])):
        return True
    return False

def palindrom_sayi_mi(liste):
    for i in liste:
        i = str(i)
        if i != i[::-1]:
            return False
    palindrom = palindrom_sirali_mi(liste)
    if palindrom:
        return True
    return False

def palindrom_sayi_hesaplama(liste):
    for i in range(liste[2]+1, liste[2]+100):
        i = str(i)
        if i == i[::-1]:
            return i

def palindrom_sirali_mi(liste):
    for i in liste:
        if i != palindrom_mu(liste, liste.index(i)):
            return False
    return True

def palindrom_mu(liste, index):
    while True:
        for i in range(liste[0]+index, liste[index]+100):
            is_prime = True
            i = str(i)
            if i != i[::-1]:
                is_prime = False
            
            if is_prime:
                return int(i)
            else:
                i = int(i) + 1

def ucgensel_dizi_mi(liste):
    for i in liste:
        ture = False
        for x in range(0, i+1):
            if x**2 + x == 2 * i:
                ture = True
                break
        if ture != True:
            return False
    ucgensel = ucgensel_sirali_mi(liste)
    if ucgensel:
        return True
    return False

def ucgensel_sirali_mi(liste):
    num = liste[0]
    for i in liste[1:]:
        if i != ucgensel_mi(num, i):
            return False
        num = i
    return True

def ucgensel_mi(num, i):
    ucgen = 0
    for x in range(0, i + 1):
            if x**2 + x == 2 * i:
                ucgen = x
    num += ucgen
    return num

def ucgensel_hesaplama(liste):
    num = liste[2]
    ucgen = 0
    for x in range(0, num + 1):
            if x**2 + x == 2 * num:
                ucgen = x
    num += ucgen + 1
    return num

def aritmetrik_dizi(liste):
    if liste[2] - liste[1] == liste[1] - liste[0]:
        return ( liste[1] - liste[0]) + liste[2]

def geometrik_dizi(liste):
    if liste[2] / liste[1] == liste[1] / liste[0]:
        return int((liste[1] / liste[0]) * liste[2])

def hepsi():
    while True:
        liste = input("Örüntünün 3 tane sayısını sırayla giriniz: ")
        s1, s2, s3 = liste.split(", ")
        liste = [int(s1), int(s2), int(s3)]

        cift_mi = cift_sayi_mi(liste)
        tek_mi = tek_sayi_mi(liste)
        asal_mi = asal_sayi_mi(liste)
        fibonacci = fibonacci_mi(liste)
        karesel_mi = karesel_sayi_dizisi(liste)
        palindrom = palindrom_sayi_mi(liste)
        ucgensel = ucgensel_dizi_mi(liste)
        aritmetrik = aritmetrik_dizi(liste)
        geometrik = geometrik_dizi(liste)
    
        print(f"Çift Sayı Dizisi: {liste[2] + 2 if cift_mi else '-'}")
        print(f"Tek Sayı Dizisi: {liste[2] + 2 if tek_mi else '-'}")
        print(f"Asal Sayı Dizisi: {asal_sayi_hesaplama(liste) if asal_mi else '-'}")
        print(f"Fibonacci Sayı Dizisi: {fibonacci_hesaplama(liste) if fibonacci else '-'}")
        print(f"Karesel Sayı Dizisi: {karesel_sayi_hesaplama(liste) if karesel_mi else '-'}")
        print(f"Uçgensel Sayı Dizisi: {ucgensel_hesaplama(liste) if ucgensel else '-'}")
        print(f"Aritmetrik Sayı Dizisi: {aritmetrik if aritmetrik else '-'}")
        print(f"Geometrik Sayı Dizisi: {geometrik if geometrik else '-'}")
        baska = input("Başka örüntü inceleyecek misiniz? [e/h] : ")
        if baska.lower() != "e":
            print("Matematik örüntüleri kapsamında çalışmamız sona ermiştir. Güle güle… :)")
            break

hepsi()
```
